```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('data/0.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
print('Image shape:', image.shape)
```

```python
# read points
from handle_json import read_json

data = read_json('annotations.json')
print(len(data))
```

```python
# 0 4 width
# 0 2 height
# 0 1 depth
```

```python
edges = [
                (0, 1), (1, 5), (5, 4), (4, 0),  # Bottom face
                (2, 3), (3, 7), (7, 6), (6, 2),  # Top face
                (0, 2), (1, 3), (5, 7), (4, 6)   # Vertical edges
            ]    
  
```

```python
for i in range(len(data)):
    image = plt.imread('data/' + str(i) + '.jpg')
    
    # Display the image
    plt.imshow(image)
    
    points = data[i]['projection']
    dims = data[i]['dimensions']
    
    # Plot edges (lines) between points based on the edges array
    for edge in edges:
        point1 = points[edge[0]]
        point2 = points[edge[1]]
        
        x1, y1 = point1[0], point1[1]
        x2, y2 = point2[0], point2[1]
        
        # Plot the line by passing the x and y coordinates as separate lists
        plt.plot([x1, x2], [y1, y2], color='g', linewidth=1)  # Red line with thickness 3
    
    # Plot points and text
    for idx, p in enumerate(points):
        plt.plot(p[0], p[1], 'bo')  # Blue circle for points
        plt.text(p[0], p[1], str(idx), fontsize=12, color='red')  # Index of the point in red
        

    
    # Show the image with lines and points
    plt.show()
    print('Image:', i)
    print('Dimensions:', dims)
```
